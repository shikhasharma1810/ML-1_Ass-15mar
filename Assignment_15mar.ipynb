```python
1. Artificial Intelligence: is technology that enables computers and machines to simulate human intelligence
and problem-solving capabilities. It is often mentioned together with machine learning and deep learning. 
These disciplines involve the development of AI algorithms, modeled after the decision-making processes of the
human brain, that can ‘learn’ from available data and make increasingly more accurate classifications or 
predictions over time.

2. Machine Learning: is the form of AI that allows computers to learn patterns and make predictions based on data,
without being explicitly programmed to do so. Some machine learning algorithms, called deep learning, can specialize
in training themselves to detect patterns.

3. Deep Learning: is the subfield of AI that teaches computers to process data in a way inspired by the human brain.
Its models can recognize complex patterns in data to produce insights and predictions.
```

```python
Q2: What is supervised learning? List some examples of supervised learning.
```

```python
Supervised learning is a machine learning technique that uses labeled data to train algorithms to recognize
patterns and predict outcomes. The algorithm learns to map the input data to the desired output, allowing it
to make predictions for new, unseen data.

Examples:
    
1. Classification: Spam filters, image recognition, fraud detection systems
2. Regression: Weather predictions, pricing predictions
3. Recommendation engines: Recommending products or content
4. Speech recognition: Voice recognition technology
5. Medical diagnosis: Predicting heart disease
6. Stock market: Predicting stock prices
```

```python
Q3:  What is unsupervised learning? List some examples of unsupervised learning.
```

```python
Unsupervised learning is a machine learning technique that uses unlabeled data and datasets of clusters
or similar groups.
Example:
Customer Segmentation- Retail companies can use unsupervised learning to group customers based on their 
purchasing patterns and behaviors. 
```

```python
Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?
```

```python
Supervised learning uses labeled data, expecting to develop predictive capability.
Unsupervised learning is discovering process, diving into unlabeled data to capture hidden information. 
Semi-supervised learning uses the concept of both supervised and unsupervised learning.
```

```python
Q6: What is train, test and validation split? Explain the importance of each term.
```

```python
Train, test, and validation split involves dividing a dataset into three subsets: training, testing,
and validation. Each term plays a crucial role in model development:
    
Training set: Used to train the model. The model learns from the data in this set.

Validation set: Used to evaluate the model's performance and fine-tune hyperparameters. The model doesn't
learn from this data, but uses it for evaluation and feedback.

Test set: Used to evaluate the final performance of the model and its ability to generalize to new, unseen data.
The model's performance on this set indicates how well it will perform in real-world applications. 
```

```python
Unsupervised learning algorithms identify normal patterns in a data sample and then detect outliers based on
the data set's natural characteristics. The model looks for data points that deviate significantly from the
majority of the data. 
```
